## **1. Load ảnh biển báo STOP và chuyển ảnh sang gray**  

```python
img = cv2.imread("Data/image.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,5))
plt.subplot(131), plt.imshow(img), plt.title('BRG image')
plt.subplot(132), plt.imshow(img_gray), plt.title('Gray image')
plt.subplot(133), plt.imshow(img_rgb), plt.title('RGB image')
plt.show()
```

## **2. Load file XML training và xác định vật thể trong hình**  

```python
cascade_model = cv2.CascadeClassifier('stop_data.xml')

Find1 = cascade_model.detectMultiScale(img_gray, minSize =(10, 15))
Find2 = cascade_model.detectMultiScale(img_gray, minSize =(30, 50))
Find3 = cascade_model.detectMultiScale(img_gray, minSize =(200, 300))

print('Vị trí detect 1', Find1)
print('Vị trí detect 2', Find2)
print('Vị trí detect 3', Find3)

img1 = img_rgb.copy()
img2 = img_rgb.copy()
img3 = img_rgb.copy()

if len(Find1) != 0:
    for (x, y, width, height) in Find1:
        cv2.rectangle(img1, (x, y), (x + height, y + width), (0, 255, 0), 5)
if len(Find2) != 0:
    for (x, y, width, height) in Find2:
        cv2.rectangle(img2, (x, y), (x + height, y + width), (0, 255, 0), 5)
if len(Find3) != 0:
    for (x, y, width, height) in Find3:
        cv2.rectangle(img3, (x, y), (x + height, y + width), (0, 255, 0), 5)

plt.figure(figsize=(15,5))
plt.subplot(131), plt.imshow(img1), plt.title('MinSize = (10, 15)')
plt.subplot(132), plt.imshow(img2), plt.title('MinSize = (30, 50)')
plt.subplot(133), plt.imshow(img3), plt.title('MinSize = (200, 300)')
plt.show()
```

## **3. Ví dụ về xác định khuôn mặt con người**  

```python
img = cv2.imread("Data/Julie.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cascade_model = cv2.CascadeClassifier('Haar Training/cascade2xml/myfacedetector.xml')

Find = cascade_model.detectMultiScale(img_gray, minSize =(10, 15))

print('Vị trí detect', Find)

if len(Find) != 0:
    for (x, y, width, height) in Find:
        cv2.rectangle(img_rgb, (x, y), (x + height, y + width), (0, 255, 0), 5)

plt.figure(figsize=(15,5))
plt.subplot(111), plt.imshow(img_rgb), plt.title('MinSize = (10, 15)')
plt.show()
```
